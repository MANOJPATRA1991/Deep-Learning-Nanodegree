In this notebook, we train an MLP to classify images from the MNIST database.

### 1. Load MNIST Database

```python
from keras.datasets import mnist
import tensorflow

# use Keras to import pre-shuffled MNIST database
(X_train, y_train), (X_test, y_test) = mnist.load_data()

print("The MNIST database has a training set of %d examples." % len(X_train))
print("The MNIST database has a test set of %d examples." % len(X_test))
```

### 2. Visualize the First Six Training Images

```python
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.cm as cm
import numpy as np

# plot first six training images
fig = plt.figure(figsize=(20,20))
for i in range(6):
    # Subplot of one row and six columns
    ax = fig.add_subplot(1, 6, i+1, xticks=[], yticks=[])
    # Show the images from index 0 to 5 from the X_train dataset
    ax.imshow(X_train[i], cmap='gray')
    # Set title of each image with the predicted values
    # from the y_train dataset from index 0 to 5
    ax.set_title(str(y_train[i]))
```

### 3. View an Image in More Detail

```python
# Any grayscale image is interpreted by the computer as a matrix
# with one entry for each image pixel

# Each image in the MNIST database is 28 pixels high and 28 pixels wide and
# so, understood by the computer as a 28 by 28 matrix 
# => 784 (28 X 28) entries as a vector => 784 input layers
# and 10 output layers

# Whit pixels encoded as 255, black pixels as 0 and gray pixels as 114

# Multi-layer perceptron (MLP) take vectors as inputs to input layers. So,
# we need to convert the images into vectors

def visualize_input(img, ax):
    ax.imshow(img, cmap='gray')
    width, height = img.shape
    thresh = img.max()/2.5
    for x in range(width):
        for y in range(height):
            ax.annotate(str(round(img[x][y],2)), xy=(y,x),
                        horizontalalignment='center',
                        verticalalignment='center',
                        color='white' if img[x][y]<thresh else 'black')

fig = plt.figure(figsize = (12,12)) 
ax = fig.add_subplot(111)
visualize_input(X_train[0], ax)
```

### 4. Rescale the Images by Dividing Every Pixel in Every Image by 255

```python
# rescale [0,255] --> [0,1] => Rescaling each pixel to have a value between 0 and 1

# Copy of the array, cast to a specified type using astype
X_train = X_train.astype('float32')/255
X_test = X_test.astype('float32')/255 
```

### 5. Encode Categorical Integer Labels Using a One-Hot Scheme

```python
from keras.utils import np_utils

# print first ten (integer-valued) training labels
print('Integer-valued labels:')
print(y_train[:10])

# one-hot encode the labels
y_train = np_utils.to_categorical(y_train, 10)
y_test = np_utils.to_categorical(y_test, 10)

# print first ten (one-hot) training labels
print('One-hot labels:')
print(y_train[:10])
```

### 6. Define the Model Architecture

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten

# define the model
model = Sequential()
# Flatten layer is used to flatten the input and appears before
# we specify the MLP. It takes the image matrix's input and 
# converts it into a vector
model.add(Flatten(input_shape=X_train.shape[1:]))

# 'relu' activation function leaves all positive values alone and
# sends all negative values to zero
model.add(Dense(512, activation='relu'))

# Dropout consists in randomly setting a fraction rate of 
# input units to 0 at each update during training time, 
# which helps prevent overfitting.
model.add(Dropout(0.2))

model.add(Dense(512, activation='relu'))

# Dropout consists in randomly setting a fraction rate of 
# input units to 0 at each update during training time, 
# which helps prevent overfitting.
model.add(Dropout(0.2))

# Output layer with 10 nodes, where each node 'i' depicts
# the probability that the image depicted is 'i'
model.add(Dense(10, activation='softmax'))

# summarize the model
model.summary()
```

### 7. Compile the Model

```python
# compile the model
model.compile(loss='categorical_crossentropy', optimizer='rmsprop', 
              metrics=['accuracy'])
```

### 8. Calculate the Classification Accuracy on the Test Set (Before Training)

```python
# evaluate test accuracy
score = model.evaluate(X_test, y_test, verbose=0)
accuracy = 100*score[1]

# print test accuracy
print('Test accuracy: %.4f%%' % accuracy)
```

### 9. Train the Model

```python
from keras.callbacks import ModelCheckpoint   

# Fit the model weights based on the TRAIN SET
# Check how the model is doing based on the VALIDATION SET
# Check the accuracy of the trained model based on the TEST SET

# train the model
# ModelCheckpoint class saves the model after each epoch
# Arguments: filepath --> where to save the model weights
#            verbose --> 0, 1 or 2
#            save_best_only --> If true, we ask to save only the weights to 
#                               to get the best accuracy on the validation set
checkpointer = ModelCheckpoint(filepath='mnist.model.best.hdf5', 
                               verbose=1, save_best_only=True)
hist = model.fit(X_train, y_train, batch_size=128, epochs=10,
          validation_split=0.2, callbacks=[checkpointer],
          verbose=1, shuffle=True)
```

### 10. Load the Model with the Best Classification Accuracy on the Validation Set

```python
# load the weights that yielded the best validation accuracy
model.load_weights('mnist.model.best.hdf5')
```

### 11. Calculate the Classification Accuracy on the Test Set

```python
# evaluate test accuracy
score = model.evaluate(X_test, y_test, verbose=0)
accuracy = 100*score[1]

# print test accuracy
print('Test accuracy: %.4f%%' % accuracy)
```

```python
# With two hidden layers of 512 layers each, we hav an accuracy of 96.3700%.
# Increasing the number of nodes from 512 to 912 for the first hidden layer has a slightly better accuracy of 96.6900%.
# Increasing the number of nodes from 512 to 912 for the second hidden layer has a slightly better accuracy of 96.8200% compared to increasing the number of nodes for the first layer.
# Increasing the number of nodes from 512 to 912 for both the hidden layers has a rather better accuracy of 97.2700%.

# Decreasing the number of nodes from 512 to 312 for the first hidden layer has a increased accuracy of 97.3900%.
# Decreasing the number of nodes from 512 to 312 for the second hidden layer has a reduced accuracy of 96.2700% compared to increasing the number of nodes for the first layer.
# Decreasing the number of nodes from 512 to 312 for both the hidden layers has a further reduced accuracy of 95.7400%.

# Thus, inreasing the number of nodes for hidden layers results in increased accuracy but the change in accuracy with increasing nodes is very slow => a case of OVERFITTING.
# And reducing the number of nodes for hidden layers results in an overall reduced accuracy which as we can see above decreases faster => a case of UNDERFITTING.

# With the removal of the dropout layers, we noticed a slight decrease in the accuracy from 96.3700% to 96.3100%. Although the change is small but a decrease in accuracy is a sign of overfitting.

```
